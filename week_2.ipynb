```python
l = ['PM2.5', 'PM10', 'NO2', 'NOx',  'CO', 'SO2','AQI', 'NO', 'NH3', 'O3', 'Benzene', 'Toluene', 'Xylene']
for i in l:
    df[i] = df.groupby("City")[i].transform(lambda x: x.fillna(x.median()))
```

```python
df.head(5)
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
df_imputed = imputer.fit_transform(df[["PM10","NOx","NH3","O3","Benzene","Toluene","Xylene"]])
```

```python
dfimputed = pd.DataFrame(df_imputed,columns=[["PM10","NOx","NH3","O3","Benzene","Toluene","Xylene"]],index=df.index)
```

```python
df[["PM10","NOx","NH3","O3","Benzene","Toluene","Xylene"]] = dfimputed
```

```python
df = pd.get_dummies(df, columns=["City"], drop_first=True)
```

```python
x = df.drop(["Date","AQI","AQI_Bucket"],axis=1)
y = df[["AQI"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =  train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```
